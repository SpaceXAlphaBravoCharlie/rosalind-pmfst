```python
def ReverseComplement(pattern):
  reverse_complement = str()
  for nucleotide in pattern:
    reverse_complement = reverse_complement + ComplementaryBase(nucleotide)
  return reverse_complement[::-1]
```

```python
def ComplementaryBase(base):
  if base == 'A':
    return 'T'
  elif base == 'C':
    return 'G'
  elif base == 'G':
    return 'C'
  else:
    return 'A'
```
